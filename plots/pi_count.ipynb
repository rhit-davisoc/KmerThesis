```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
import re
import math
```

```python
ref_seq = "../standard_test_data/fasta_files/standard_individual.fa"

with open(ref_seq) as file:
    data = file.read().replace('\n','').replace('>n0','')

data = re.sub(r'>n.',',',str(data))
```

```python
data
```

```python
arr = data.split(",")

pi = 0

nucleotides = len(arr[0])
indvs = len(arr)

combos = math.comb(indvs,2)

for i in range(0,nucleotides):
    ref = arr[0][i]
    c_ref = 1
    c_alt = 0
    for k in range (1,indvs):
        if(arr[k][i]==ref):
            c_ref += 1
        else:
            c_alt += 1
    
    pi += (c_ref*c_alt / combos)

pi = pi / nucleotides

print(float(pi))
```
